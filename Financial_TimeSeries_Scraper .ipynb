```python
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
ticker_symbol = 'DB'

# Fetch data using yfinance's Ticker object
ticker = yf.Ticker(ticker_symbol)

data = ticker.history(start='2000-12-01', end='2023-12-01', interval='1mo')
```

```python
data
```

```python
end_of_month_data = data.resample('M').last()

# Display the first few rows of the data
print(end_of_month_data.head())
```

```python
close_data=end_of_month_data.loc[:,'Close']
close_data
```

```python
output_file = "ClosingPrice.xlsx"  # Specify the name of the output file
close_data.to_excel(output_file, index=False)  # Export without the index
```
